```python


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline3
from sklearn.preprocessing import StandardScaler, MinMaxScaler, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error


# Numerical transformer
numerical_transformer = Pipeline3([
    ('scaler', MinMaxScaler())
])

# Frequency encoding function
def frequency_encoder(X):
    X = X.copy()  # Ensure original data is not modified
    for col in X.columns:
        freq = X[col].value_counts()
        X[col] = X[col].map(freq)
    return X

# Frequency transformer
frequency_transformer = FunctionTransformer(frequency_encoder)

# Categorical transformer with frequency encoding
categorical_transformer_frequency = Pipeline([
    ('frequency_encoder', frequency_transformer),
    ('scaler', MinMaxScaler())
])

# Categorical transformer with one-hot encoding
categorical_transformer_ohe = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False)),
    ('scaler', MinMaxScaler())
])

# Preprocessor3
preprocessor3 = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, ['year', 'odometer']),
        ('cat_freq', categorical_transformer_frequency, ['manufacturer', 'model', 'state']),
        ('cat_ohe', categorical_transformer_ohe, ['fuel', 'title_status', 'transmission', 'type', 'paint_color'])
    ], remainder='passthrough'
)

# Pipeline3
pipeline3 = Pipeline(steps=[('preprocessor3', preprocessor3),
                           ('regressor', RandomForestRegressor(random_state=42))])

# Fit the pipeline3
pipeline3.fit(x_train, y_train)

# Make predictions
y_pred3 = pipeline3.predict(x_test)



import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score

mae = mean_absolute_error(y_test, y_pred3)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred3)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# R² Score
r2 = r2_score(y_test, y_pred3)

print("Regression Metrics:")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```
